# Calling large language models with LangChain

This notebook demonstrates how to configure and invoke large language models (LLMs) through LangChain. Each section focuses on a different model provider and follows the same basic workflow:

1. Load credentials from the local environment.
2. Create a LangChain chat model.
3. Invoke the model with a question.
4. Display the response.

Keep API keys in a local `.env` file. Never place credentials directly in a notebook.

## Environment setup

The project uses `python-dotenv` to load `GOOGLE_API_KEY` and `GROQ_API_KEY` from `.env`. The provider integrations come from `langchain-google-genai` and `langchain-groq`.

```python
import os
import sys

from dotenv import load_dotenv

load_dotenv()

assert os.getenv("GOOGLE_API_KEY"), "Add GOOGLE_API_KEY to your .env file."
assert os.getenv("GROQ_API_KEY"), "Add GROQ_API_KEY to your .env file."
print(f"Python executable: {sys.executable}")
```

## 1. Google Gemma 4

This example uses Google's instruction-tuned `gemma-4-31b-it` model. A temperature of `0` reduces randomness, which is useful when asking a factual question.

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemma-4-31b-it",
    temperature=0,
)

llm
```

### Invoke the model

Chat models accept a sequence of messages. The system message defines how the model should respond, while the human message contains the user's question.

```python
from IPython.display import Markdown, display
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(
        content=(
            "You are a concise astronomy educator. "
            "Answer factual questions clearly in two or three sentences."
        )
    ),
    HumanMessage(
        content="How many moons does Mars have, and what are their names?"
    ),
]

response = llm.invoke(messages)
display(Markdown(response.text))
```

## 2. Google Gemini 2.5 Flash

`gemini-2.5-flash` is a fast, general-purpose Gemini model. The same LangChain interface works across Google models, so only the model identifier and variable names need to change.

```python
gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
)

gemini_llm
```

### Invoke the model

This example again separates behavioral instructions from the user's question by using system and human messages.

```python
gemini_response = gemini_llm.invoke(messages)
display(Markdown(gemini_response.text))
```

## 3. Groq Qwen3-32B

Groq provides fast hosted inference for `qwen/qwen3-32b`. LangChain exposes it through the `ChatGroq` integration.

```python
from langchain_groq import ChatGroq

qwen_llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    reasoning_format="hidden",
)

qwen_llm
```

### Invoke the model

The same system and human messages are passed to Qwen3-32B so its response can be compared directly with the Google models.

```python
qwen_response = qwen_llm.invoke(messages)
display(Markdown(qwen_response.text))
```
